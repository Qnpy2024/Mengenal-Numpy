```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv('/content/drive/MyDrive/2024 1 Visualisasi Data & Analisis Eksplorasi Data/Salary_Data.csv')
data
```

```python
data = data.dropna()
data
```

```python
data.isnull().sum()
```

```python
x = data.iloc[:,4]
x.head(10)
```

```python
y = data.iloc[:,-1]
y.head(5)
```

```python
plt.scatter(x,y)
plt.title('Experience vs Salary')
plt.xlabel('Experience')
plt.ylabel('Salary')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33,random_state=51)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
linear = LinearRegression()
linear.fit(np.array(x_train).reshape(-1,1),y_train)
```

```python
linear.coef_
```

```python
linear.intercept_
```

```python
Y_pred = linear.predict(np.array(x_test).reshape(-1,1))
```

```python
Y_pred
```

```python
y_test
```

```python
score = r2_score(y_test,Y_pred)
score
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,Y_pred,color='red')
plt.title('Experience vs Salary')
plt.xlabel('Experience')
plt.ylabel('Salary')
```
